EvalTraj.py is used to simulate the force field. Below we show how the force field changes depending on three parameters: time lag between mouse and distractor position, relative contribution of the circle force and the orientation displacement. We show the measures - the distance from the nearest force minimum (which is unfortunately rather noisy) and the magnitude of the net force at the mouse location.

```python
def plotR(x,R,g=0):
    a=[]
    for h in range(R.shape[0]):
        a.append([])
        for i in range(R.shape[1]): 
            for j in range(R.shape[2]):
                for d in range(R.shape[3]):
                    dd=R[h,i,j,d,:,g]
                    if g==2: sel=R[h,i,j,d,:,-1]<11.73
                    else: sel=dd<np.inf
                    a[-1].append(stats.nanmedian(dd[sel]))
        #plt.plot(x,a[-1])
    plt.plot(x,np.array(a).mean(0),'k',lw=3)
for f in [1]:
    plt.figure(figsize=(14,9))
    for g in range(2):
        suf=['n','e'][f]
        plt.subplot(2,3,g*3+1)
        R=np.load('Rfms%s.npy'%suf)
        x=np.load('fms.npy')
        try:plotR(x,R,g=[0,2][g])
        except: pass
        plt.ylabel(['locmin distance','abs force'][g])
        plt.xlabel('time shift')
        plt.subplot(2,3,g*3+2)
        R=np.load('Rcms%s.npy'%suf)
        plotR(np.load('cms.npy'),R,g=[0,2][g])
        plt.xlabel('circle mass')
        plt.subplot(2,3,g*3+3)
        R=np.load('Rdps%s.npy'%suf)
        plotR(np.load('dps.npy'),R,g=[0,2][g])
        plt.xlabel('displacement')
```
