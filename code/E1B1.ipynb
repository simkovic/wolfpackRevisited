## Location Recall

We start with the data from the location judgment task. We have data from subjects 301 to 313.

```python
vpn=range(301,314)
D=loadDataB12(vpn)
print D.shape
```

We have loaded all interesting data in the array D. The last dimension of D gives for each subject, block and trial the following information 
<ul>
	<li>mean proportion of time spent in wolfpack quadrants [0,1]</li>
	<li>distance travelled during a trial in degrees per second</li>
	<li>whether the judgment task targeted a wolf (1) or perpendicular (0)</li>
    <li>proportion of time spent in collision with agents [0,1]</li>
    <li>whether the location recall task data from this trial should be included in the analysis (1-yes,0-no) </li>
    <li>magnitude of the displacement in the location recall task in degrees</li>
    <li>direction of the displacement in radians</li>
    <li>direction of the target's motion in radians</li>
    <li>direction of the green circle w.r.t. to the target in radians</li>
    <li>orientation of the nose in radians</li>
    <li>magnitude of manipulated displacement in block 2</li>
</ul>

Trials where the subject clicked on the green agent as well as trials where the magnitude of displacement is larger than 4 degrees are excluded.

Let's look at how the different factors affect displacement.

```python
plotB1(D)
```

Chasee Gravity and Orientation Displacement show strong effects however the two factors are correlated. We therefore use linear regression to estimate the magnitude of these two effects for each subject.

We expect four predictors to influence the displacement. 

<ol>
	<li> <strong>Representational gravity</strong> pulls objects down along the vertical axis </li> 
    <li>Under the influence of <strong>representational momentum</strong> subjects interpolate future positions along the motion trajectory of object and recall these as its last postion.</li> 
    <li>Third, we call the displacement in the direction of the circles orientation (as indicated by the eyes' location or physical displacement in the third block), <strong>orientation displacement</strong>.</li>
    <li>Finally, we expect that the last position of the green circle (controled by mouse) will influence the judgment. In particular we expect the judgment to be pulled towards the last mouse position. We call this effect the <strong>lazy-hand gravity</strong>.</li>
</ol>
More formally, let $\mathbf \theta_i =[\theta_{ix}, \theta_{iy}]$ be the vector of the difference between the indicated and the actual position of the vanished circle at trial $i$ along the two axes. Let $\phi_{ik}$ be the angle given by the direction of the predictor at trial $i$, with $k$ giving the indexing the predictor (as listed in the previous paragraph). Then we are interested in the estimates of the regression coefficients $\alpha_k$ given by the equation

$$
\mathbf \theta_i \sim  \mathcal{N}\left(\sum_k \alpha_k \begin{bmatrix} \cos \phi_{ik} \\ \sin \phi_{ik} \end{bmatrix}, I\sigma_\theta \right)
$$

We first compute the predictor and outcome variables.

```python
G=[[],[],[],[]];sels=[]
for i in range(len(vpn)):
    # predictor var
    px=np.cos([np.zeros(T),D[i,0,:,7],D[i,0,:,8],D[i,0,:,9]])
    py=np.sin([np.ones(T)*np.pi/2.0,D[i,0,:,7],D[i,0,:,8],D[i,0,:,9]])
    
    # outcome var
    ox=np.cos(D[i,0,:,6])*D[i,0,:,5]
    oy=np.sin(D[i,0,:,6])*D[i,0,:,5]
    sel=D[i,0,:,4]==1
    sels.append(sel)
    G[0].append(px.T);G[1].append(py.T)
    G[2].append(ox); G[3].append(oy)
sels=np.array(sels)
seln=sels.mean(1)>0.1
sels=sels[seln,:]
for i in range(seln.size):
    if not seln[i]: 
        for k in range(4): G[k].pop(i)
```

```python
model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real<lower=-1,upper=1> px[N,T,4]; // predictor on x axis
    real<lower=-1,upper=1> py[N,T,4]; // predictor on y axis
    real<lower=-20,upper=20> ox[N,T]; // outcome on x axis
    real<lower=-20,upper=20> oy[N,T]; // outcome on y axis
    int<lower=0,upper=1> sel[N,T]; // indicator var to include item
}
parameters {
    real a1[N];
    real a2[N];
    real a3[N];
    real a4[N];
    real<lower=0> sigma[N];
}
model {
    real mx;
    real my;
    for (n in 1:N){
        for (t in 1:T){
            if (sel[n,t]==1){
                mx<-a2[n]*px[n,t,2]+a3[n]*px[n,t,3]+a4[n]*px[n,t,4];
                my<-a1[n]*py[n,t,1]+a2[n]*py[n,t,2]+a3[n]*py[n,t,3]+a4[n]*py[n,t,4];
                ox[n,t] ~ normal(mx,sigma[n]);
                oy[n,t] ~ normal(my,sigma[n]);
            }
        }
    }
}
"""
dat = {'N': seln.sum(),'T':T,'sel':np.int32(sels),'px': G[0],'py':G[1],'ox':G[2],'oy':G[3]}
fit = pystan.stan(model_code=model, data=dat,iter=1000, chains=4)
print fit
pystanErrorbar(fit)
saveStanFit(fit,'E1B1L.fit')
```

The box plot whiskers are wide in all figures. Let's estimate the population values with a hierarchical model. From the above plots it looks plausible to assume normal distribution for subject parameters. Notable exception are the Orientation displacement parameters. We see two groups of subjects - those who compensate and those who do not. Nevertheless confidence intervals are wide and show considerable overlap.

```python

model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real<lower=-1,upper=1> px[N,T,4]; // predictor on x axis
    real<lower=-1,upper=1> py[N,T,4]; // predictor on y axis
    real<lower=-20,upper=20> ox[N,T]; // outcome on x axis
    real<lower=-20,upper=20> oy[N,T]; // outcome on y axis
    int<lower=0,upper=1> sel[N,T]; // indicator var to include item
}
parameters {
    real a1[N];
    real a2[N];
    real a3[N];
    real a4[N];
    real<lower=0,upper=10> sigma[N];
    real ma[4];
    real<lower=0> sa[6];
}

model {
    int ind[N];
    real mx;
    real my;
    real ps[2];
    ma~normal(0,0.5);
    a1~normal(ma[1],sa[1]);
    a2~normal(ma[2],sa[2]);
    a3~normal(ma[3],sa[3]);
    a4~normal(ma[4],sa[4]);
    sigma~normal(sa[5],sa[6]);
    for (n in 1:N){
        for (t in 1:T){
            if (sel[n,t]==1){
                mx<-a2[n]*px[n,t,2]+a3[n]*px[n,t,3]+a4[n]*px[n,t,4];
                my<-a1[n]*py[n,t,1]+a2[n]*py[n,t,2]+a3[n]*py[n,t,3]+a4[n]*py[n,t,4];
                ox[n,t] ~ normal(mx,sigma[n]);
                oy[n,t] ~ normal(my,sigma[n]);
            }
        }
    }
}
"""
fit = pystan.stan(model_code=model, data=dat,iter=15000, chains=4,warmup=2000,thin=5)
print fit
pystanErrorbar(fit)
saveStanFit(fit,'E1B1LH.fit')
```

## Leave Me Alone Task

Let's now look at the data from leave me alone task. We will start with analysis similar to the one done by Gao and colleagues

```python

# do t-test a la gao10
s=D[:,0,:,0].mean(1)
%load_ext rmagic 
%Rpush s
%R res=t.test(s,mu=0.5,alternative="less")
%Rpull res
print 'p-value ',res[2]
```

Red line shows the sample mean (n=7) reported by Gao. Green surface shows the reported confidence interval. Blue points are the values of subjects in our sample. Our subjects do not show the expected avoidance of wolfpack. In fact all but one observed values falls outside of the reported CI. 

The subjects do not show the expected avoidance of wolfpack areas but maybe in our sample this behavior is modulated by other variables. Let's first look whether the magnitude of avoidance correlates with the magnitude of the <strong>Orientation Displacement</strong>.

```python
plt.figure()
plt.scatter(np.median(w['a4'],axis=0),D[seln,0,:,0].mean(1))
plt.ylim([0.45,0.55])
plt.plot([-0.05,0.3],[0.5,0.5],'k')
plt.xlim([-0.05,0.3])
plt.xlabel('Orientation displacement')
plt.ylabel('Mean proportion of trial spent in wolfpack quadrants')
```

Ok, Orientation Displacement doesn't seem to matter with respect to avoidance of wolfpack quadrants. Let's look at properties of the mouse movement and whether differences between subjects could help us explain the replication failure.
```python
plt.scatter(D[:,0,:,1].flatten(), D[:,0,:,0].flatten())
plt.xlabel('Average mouse speed')
plt.ylabel('Proportion of trial spent in wolfpack quadrants')
plt.title('Correlation across trials')
np.corrcoef(D[:,0,:,1].flatten(), D[:,0,:,0].flatten())[0,1]

plt.figure()
plt.scatter(D[:,0,:,1].mean(1), D[:,0,:,0].mean(1))
plt.xlabel('Mean average mouse speed')
plt.ylabel('Mean proportion of trial spent in wolfpack quadrants')
plt.title('Correlation across subjects')
np.corrcoef(D[:,0,:,1].mean(1), D[:,0,:,0].mean(1))[0,1]
```

We start by exploring a truncated gaussian.

```python
G= np.copy(D[:,0,:,0])
G[G==1]=0.999
G[G==0]=0.001
dat = {'N': G.shape[0],'T':T,'owt':G}
ipars=[{'mu': G.mean(1),'mmu':0.5}]*4
```

```python
model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real<lower=0,upper=1> owt[N,T]; // proportion of time spent in wolfpack areas
}
parameters {
    real<lower=0,upper=1> mu[N];
    real<lower=0,upper=1> si[N];
    real<lower=0,upper=1> mmu ;
    real<lower=0,upper=1> smu;
    real<lower=0,upper=1> pmu;
    real<lower=0,upper=1> psi;
    real<lower=0,upper=1> powt[T];
}
model {
    for (n in 1:N){
        mu[n]~normal(mmu,smu)T[0,1];
        for (t in 1:T){
            owt[n,t]~normal(mu[n],si[n])T[0,1];   
        }
    }
    pmu~normal(mmu,smu)T[0,1];
    for (t in 1:T){
        powt[t]~normal(pmu,psi)T[0,1];  
    }
}
"""
fit = pystan.stan(model_code=model, data=dat,iter=5000, chains=4,init=ipars,warmup=1000)
print fit
pystanErrorbar(fit)
saveStanFit(fit,'E1B1WGG.fit')
check(fit.extract()['powt'])
```

We see problems with fitting the data of subject 12. The gaussian does not fit the flat distribution of this subject (see histograms of the actual data). The standard deviation does not fit in the range [0,1]. However if we choose larger variance we will have problems with reliably fitting the mean since gaussian with large variance becomes flat in region [0,1] 

Furthermore the distribution of min and max values peakes at the far edges, while in the data the distribution peaked at 0.1 and 0.9 respectively.  

Let's try hierarchical model with beta distribution instead of gaussian.

```python

model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real<lower=0,upper=1> owt[N,T]; // proportion of time spent in wolfpack areas
}
parameters {
    real<lower=0,upper=1> mu[N];
    real<lower=0,upper=100> nu[N];
    real<lower=0,upper=1> mmu ;
    real<lower=0,upper=1000> nmu;
    real<lower=0,upper=1000> ga;
    real<lower=0,upper=1000> gb;
    
    real<lower=0,upper=1> pmu;
    real<lower=0,upper=100> pnu;
    real<lower=0,upper=1> powt[T];
}
model {
    nmu~pareto(0.1,1.5);
    nu~gamma(ga,gb);
    mu~beta(mmu*nmu,(1-mmu)*nmu);
    for (n in 1:N){
        for (t in 1:T){
            owt[n,t]~beta(mu[n]*nu[n],(1-mu[n])*nu[n]);
        }
    }
    pnu~gamma(ga,gb);
    pmu~beta(mmu*nmu,(1-mmu)*nmu);
    powt~beta(pmu*pnu,(1-pmu)*pnu);
}
"""
fit = pystan.stan(model_code=model, data=dat,iter=5000, chains=4,init=ipars,warmup=1000)
print fit
pystanErrorbar(fit)
saveStanFit(fit,'E1B1WBB.fit')
```

```python
check(fit.extract()['powt'])
```

We now explore a combination of beta at trial level and gaussian at subject level. This should work since subject mean proportions are centered around 0.5.

```python
model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real<lower=0,upper=1> owt[N,T]; // proportion of time spent in wolfpack areas
}
parameters {
    real<lower=0,upper=1> mu[N];
    real<lower=1,upper=1000> nu[N];
    real<lower=0,upper=1> mmu ;
    real<lower=0,upper=1> smu;
    real<lower=0,upper=1000> a;
    real<lower=0,upper=1000> b;
    
    real<lower=0,upper=1> pmu;
    real<lower=1,upper=1000> pnu;
    real<lower=0,upper=1> powt[T];
}
model {
    
    nu~gamma(a,b);
    for (n in 1:N){
        mu[n]~normal(mmu,smu) T[0,1];
        for (t in 1:T){
            owt[n,t]~beta(mu[n]*nu[n],(1-mu[n])*nu[n]);
        }
    }
    pmu~normal(mmu,smu)T[0,1];
    pnu~gamma(a,b);
    powt~beta(pmu*pnu,(1-pmu)*pnu); 
}
"""
fit = pystan.stan(model_code=model, data=dat,iter=5000, chains=4,init=ipars,thin=5,warmup=1000)
print fit
pystanErrorbar(fit)
saveStanFit(fit,'E1B1WGB.fit')
check(fit.extract()['powt'])
```

Finally we look at normal model fit to logit transformed data.

```python

model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real owt[N,T]; // proportion of time spent in wolfpack areas
}
parameters {
    real<lower=-7,upper=7> mu[N];
    real<lower=0,upper=10> si[N];
    real<lower=-7,upper=7> mmu ;
    real<lower=0,upper=10> smu;
    real<lower=0,upper=1000> a;
    real<lower=0,upper=1000> b;
    
    real<lower=0> psi;
    real<lower=-7,upper=7> pmu;
    real powt[T];
}
transformed parameters{
    real tpowt[T];
    real tmu[N];
    real tsi[N];
    real tmmu ;
    real tsmu;
    for (n in 1:N){
        tmu[n]<-inv_logit(mu[n]);
        tsi[n]<-inv_logit(si[n]);
    }
    tmmu<-inv_logit(mmu);
    tsmu<-inv_logit(smu);
    for (t in 1:T) tpowt[t]<-inv_logit(powt[t]);

}
model {
    si~gamma(a,b);
    for (n in 1:N){
        mu[n]~normal(mmu,smu);
        for (t in 1:T){
            owt[n,t]~normal(mu[n],si[n]);
        }
    }
    pmu~normal(mmu,smu);
    psi~gamma(a,b);
    powt~normal(pmu,psi);
}
"""
G= np.log(G)-np.log(1-G)
dat = {'N': D.shape[0],'T':T,'owt':G}
ipars=[{'mu': G.mean(1),'mmu':0}]*4
fit = pystan.stan(model_code=model, data=dat,iter=5000, chains=4,init=ipars,warmup=1000)
saveStanFit(fit,'E1B1WGI.fit')
pystanErrorbar(fit)
check(fit.extract()['tpowt'])
```

The gaussian at the subject level doesn't fit the data well. The beta model and the logit-transformed gaussian do much better job. We will use the beta model since the magnitude of the coefficient is easier to interpret.
