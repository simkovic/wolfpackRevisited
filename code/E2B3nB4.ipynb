We first estimate a separate model for each condition

```python
indat=[]

man=[-0.4,-0.2,0,0.2,0.4]
for m in man:
    if m==0: 
        d=np.copy(B3)
        seln=~np.all(np.isnan(d[:,0,:]),1)
        d=d[seln,:,:]
    else:
        d=[]
        for i in range(B4.shape[0]):
            temp=np.copy(B4[i,:,:])
            temp[:,temp[2,:]!=m]=np.nan
            ind=np.argsort(temp[2,:])
            temp=temp[:,ind]
            if np.any(~np.isnan(temp[0,:])):
                d.append(temp[:,:12])
        d=np.array(d)
    print d.shape
    indat.append({'N':d.shape[0],'T':d.shape[2],'dr':d[:,:2,:]})
```

```python
model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real dr[N,2,T]; // discrepancy
}
parameters {
    real<lower=-10,upper=10> mu[N,2];
    real<lower=0,upper=10> sigma[N,2];
    real<lower=-10,upper=10> mmu[2];
    real<lower=0,upper=10> smu[2];
}
model {  
    for (n in 1:N){
        mu[n,1]~normal(mmu[1],smu[1]);
        mu[n,2]~normal(mmu[2],smu[2]);
        for (t in 1:T){
            dr[n,1,t] ~ normal(mu[n,1],sigma[n,1]);
            dr[n,2,t] ~ normal(mu[n,2],sigma[n,2]);
        }
    }
}
"""
for m in range(len(man)):
    fit = pystan.stan(model_code=model, data=indat[m],iter=5000,chains=4,thin=1)
    saveStanFit(fit,'E2B3BH%d.fit'%m)
    print fit
    printCI(fit.extract(),'mmu')
```

```python
model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real drr[N,2,T]; // discrepancy
    real mn[N,T]; // manipulated value
}
parameters {
    real mu[N,2];
    real<lower=0,upper=10> sigma[N,2];
    real r[N,2];
    real<lower=-10,upper=10> mmu[2];
    real<lower=0,upper=10> smu[2];
    real mr[2];
    real<lower=0> sr[2];
}
model {
    
    for (n in 1:N){
        mu[n,1]~normal(mmu[1],smu[1]);
        mu[n,2]~normal(mmu[2],smu[2]);
        r[n,1]~normal(mr[1],sr[1]);
        r[n,2]~normal(mr[2],sr[2]);
        for (t in 1:T){
            drr[n,1,t] ~ normal(mu[n,1]+r[n,1]*mn[n,t],sigma[n,1]);
            drr[n,2,t] ~ normal(mu[n,2]+r[n,2]*mn[n,t],sigma[n,2]);
        }
    }
}
"""
seln= ~np.isnan(B4[:,0,0])
D=np.concatenate([B3,B4],2)
D=D[seln,:,:]
dat = {'N': D.shape[0],'T':D.shape[2],'drr':D[:,:2,:],'mn':D[:,2,:]}
fit = pystan.stan(model_code=model, data=dat,iter=5000, chains=4,thin=2,warmup=1000)
print fit
plotB3fit(fit,pars=['r','mr'],suffix='B4aHNr')
plt.figure()
plotB3fit(fit,pars=['mu','mmu'],suffix='B4aHNm')
saveStanFit(fit,'E2B3BHreg.fit')
```
