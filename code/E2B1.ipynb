We now move to estimation. We first estimate a separate model for each subject.

```python
G=[[],[],[],[]];sels=[]
for i in range(len(vpn)):
    # predictor var
    px=np.cos([np.zeros(T),D[i,0,:,7],D[i,0,:,8],D[i,0,:,9]])
    py=np.sin([np.ones(T)*np.pi/2.0,D[i,0,:,7],D[i,0,:,8],D[i,0,:,9]])
    
    # outcome var
    ox=np.cos(D[i,0,:,6])*D[i,0,:,5]
    oy=np.sin(D[i,0,:,6])*D[i,0,:,5]
    sel=np.logical_and(D[i,0,:,4]==1,~np.isnan(D[i,0,:,4]))
    
    if sel.mean()>0.1:
        sels.append(sel)
        G[0].append(px.T);G[1].append(py.T)
        G[2].append(ox); G[3].append(oy)
sels=np.array(sels)
dat = {'N': sels.shape[0],'T':T,'sel':np.int32(sels),'px': G[0],'py':G[1],'ox':G[2],'oy':G[3]}
```

```python
model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real px[N,T,4]; // predictor on x axis
    real py[N,T,4]; // predictor on y axis
    real ox[N,T]; // outcome on x axis
    real oy[N,T]; // outcome on y axis
    int<lower=0,upper=1> sel[N,T]; // indicator var to include item
}
parameters {
    real a1[N];
    real a2[N];
    real a3[N];
    real a4[N];
    real<lower=0> sigma[N];
}
model {
    real mx;
    real my;
    for (n in 1:N){
        for (t in 1:T){
            if (sel[n,t]==1){
                mx<-a2[n]*px[n,t,2]+a3[n]*px[n,t,3]+a4[n]*px[n,t,4];
                my<-a1[n]*py[n,t,1]+a2[n]*py[n,t,2]+a3[n]*py[n,t,3]+a4[n]*py[n,t,4];
                ox[n,t] ~ normal(mx,sigma[n]);
                oy[n,t] ~ normal(my,sigma[n]);
            }
        }
    }
}
"""

fit = pystan.stan(model_code=model, data=dat,iter=1000, chains=4)
print fit
pystanErrorbar(fit)
saveStanFit(fit,'E2B1L.fit')
```

We remove the predictor for representational gravity, since for most subjects it is not significant and runs in the opposite direction, which is implausible.

We now estimate the hierarchical model.

```python
modelhier = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real px[N,T,4]; // predictor on x axis
    real py[N,T,4]; // predictor on y axis
    real ox[N,T]; // outcome on x axis
    real oy[N,T]; // outcome on y axis
    int<lower=0,upper=1> sel[N,T]; // indicator var to include item
}
parameters {
    real a2[N];
    real a3[N];
    real a4[N];
    real<lower=0,upper=10> sigma[N];
    real ma2;
    real ma3;
    real ma4;
    real<lower=0> sa[5];
}


model {
    int ind[N];
    real mx;
    real my;
    ma2~normal(0,2);ma3~normal(0,2);ma4~normal(0,2);
    a2~normal(ma2,sa[2]);
    a3~normal(ma3,sa[3]);
    a4~normal(ma4,sa[4]);
    sigma~gamma(sa[1],sa[5]);
    for (n in 1:N){
        for (t in 1:T){
            if (sel[n,t]==1){
                mx<-a2[n]*px[n,t,2]+a3[n]*px[n,t,3]+a4[n]*px[n,t,4];
                my<-a2[n]*py[n,t,2]+a3[n]*py[n,t,3]+a4[n]*py[n,t,4];
                ox[n,t] ~ normal(mx,sigma[n]);
                oy[n,t] ~ normal(my,sigma[n]);
            }
        }
    }
}
"""
fit = pystan.stan(model_code=modelhier, data=dat,iter=5000, chains=4,thin=1)
print fit
pystanErrorbar(fit)
saveStanFit(fit,'E2B1LH.fit')
```

The results look reasonable. We see strong influence of lazy hand gravity. The orientation displacement of darts is smaller than with bugs, although this difference is not reliable. Nevertheless the dart position recall is inconsistent with localization of the dart's center at the concave vertex.

```python
def check(d):
    plt.figure(figsize=(16,10))
    for i in range(min(d.shape[0],32)):
        plt.subplot(5,8,i+1)
        plt.hist(d[i,:],np.linspace(0,1,21))
        plt.ylim([0,20])
        plt.title('ID %d'%i)
        plt.gca().set_xticklabels([])
    plt.figure(figsize=(15,6))
    J=[np.mean(d,1),np.median(d,1),np.min(d,1),np.max(d,1),d.std(1),sap(d.T,75)-sap(d.T,25), np.max(d,1)-np.min(d,1) ]
    tts=['mean','median','min','max','standard deviation','IQR','range']
    for k in range(len(J)):
        plt.subplot(2,4,k+1);plt.ylim([0,d.shape[0]/2])
        plt.hist(J[k],np.linspace(0,1,21))
        plt.title(tts[k])
check(D[:,0,:,0])
```

The data are more peeked than in Experiment 1, which is probably because subject move much more.

We explore two models which provided the best fit in Experiment 1. We start with beta at trial level and gaussian at subject level.

```python
G= np.copy(D[:,0,:,0])
G[G==1]=0.999
G[G==0]=0.001
sel= ~np.isnan(G)
seln = np.any(sel,1)
sel=np.int32(sel[seln,:])
G=G[seln,:]
dat = {'N': G.shape[0],'T':T,'owt':G,'sel':sel}
ipars=[{'mu': G.mean(1),'mmu':0.5}]*4
```

```python
model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real owt[N,T]; // proportion of time spent in wolfpack areas
    int sel[N,T];
}
parameters {
    real<lower=0,upper=1> mu[N];
    real<lower=0,upper=1000> nu[N];
    real<lower=0,upper=1> mmu ;
    real<lower=0,upper=1> smu;
    real<lower=0,upper=1000> a;
    real<lower=0,upper=1000> b;
    
    real<lower=0,upper=1> pmu;
    real<lower=0,upper=1000> pnu;
    real<lower=0,upper=1> powt[T];
}
model {
    
    nu~gamma(a,b);
    for (n in 1:N){
        mu[n]~normal(mmu,smu) T[0,1];
        for (t in 1:T){
            if (sel[n,t]==1) owt[n,t]~beta(mu[n]*nu[n],(1-mu[n])*nu[n]);
        }
    }
    pmu~normal(mmu,smu)T[0,1];
    pnu~gamma(a,b);
    powt~beta(pmu*pnu,(1-pmu)*pnu); 
}
"""
fit = pystan.stan(model_code=model, data=dat,iter=5000, chains=4,init=ipars,thin=5,warmup=1000)
print fit
pystanErrorbar(fit)
saveStanFit(fit,'E2B1WGB.fit')
check(fit.extract()['powt'])
```

```python
model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real owt[N,T]; // proportion of time spent in wolfpack areas
    int sel[N,T];
}
parameters {
    real<lower=0,upper=1> mu[N];
    real<lower=0,upper=1000> nu[N];
    real<lower=0,upper=1> mmu;
    real<lower=0,upper=1000> nmu;
    real<lower=0,upper=1000> a;
    real<lower=0,upper=1000> b;
    
    real<lower=0,upper=1> pmu;
    real<lower=0,upper=1000> pnu;
    real<lower=0,upper=1> powt[T];
}
model {
    nmu~pareto(0.1,1.5);
    nu~gamma(a,b);
    for (n in 1:N){
        mu[n]~beta(mmu*nmu,(1-mmu)*nmu);
        for (t in 1:T){
            if (sel[n,t]==1) owt[n,t]~beta(mu[n]*nu[n],(1-mu[n])*nu[n]);
        }
    }
    pmu~beta(mmu*nmu,(1-mmu)*nmu);
    pnu~gamma(a,b);
    powt~beta(pmu*pnu,(1-pmu)*pnu); 
}
"""
fit = pystan.stan(model_code=model, data=dat,iter=5000, chains=4,init=ipars,thin=5,warmup=1000)
print fit
pystanErrorbar(fit)
saveStanFit(fit,'E2B1WBB.fit')
check(fit.extract()['powt'])
```

```python

model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real owt[N,T]; // proportion of time spent in wolfpack areas
    int sel[N,T]; 
}
parameters {
    real<lower=-7,upper=7> mu[N];
    real<lower=0,upper=10> si[N];
    real<lower=-7,upper=7> mmu ;
    real<lower=0,upper=10> smu;
    real<lower=0,upper=1000> a;
    real<lower=0,upper=1000> b;
    
    real<lower=0> psi;
    real<lower=-7,upper=7> pmu;
    real powt[T];
}
transformed parameters{
    real tpowt[T];
    real tmu[N];
    real tmmu ;
    for (n in 1:N){
        tmu[n]<-inv_logit(mu[n]);
    }
    tmmu<-inv_logit(mmu);
    for (t in 1:T) tpowt[t]<-inv_logit(powt[t]);

}
model {
    si~gamma(a,b);
    for (n in 1:N){
        mu[n]~normal(mmu,smu);
        for (t in 1:T){
            if (sel[n,t]==1) owt[n,t]~normal(mu[n],si[n]);
        }
    }
    pmu~normal(mmu,smu);
    psi~gamma(a,b);
    powt~normal(pmu,psi);
}
"""
G= np.log(G)-np.log(1-G)
dat = {'N': G.shape[0],'T':T,'owt':G,'sel':sel}
ipars=[{'mu': G.mean(1),'mmu':0}]*4
fit = pystan.stan(model_code=model, data=dat,iter=5000, chains=4,init=ipars,warmup=1000)
saveStanFit(fit,'E2B1WGI.fit')
pystanErrorbar(fit)
check(fit.extract()['tpowt'])
```
