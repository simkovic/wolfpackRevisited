The estimation is not straight-forward because we have so much data with considerable variability. If we use beta parametrized by mean proportion $\mu$ and sample size $N$ distribution to model trial variability we have the following problem. Some subjects show sharply peaked distributions i.e. $N$ goes to some large number. This causes problems for the sampler. If we use gamma prior, it takes eternity to converge the high $N$. If we use flat prior with wide range the remaining $N$s (which are small around 10) are difficult to find.

We can try truncated gaussian instead, but we get an opposite and more severe problem. We have some subjects with very flat distribution so that the variance of the gaussian is pulled to infinity and mean becomes practically unidentifiable (because the gaussian is flat in the region [0,1]). 

Our current best solution is to use a constrained flat prior [0,1000] or [0,100] for $N$. This prior doesn't allow $N$ (the variable nmu below in particular) to reach its true posterior, but we think it is justified because a) higher values are implausible and b) the absolute difference in terms of the variance $\sigma= \mu*(1-\mu)/(1+N)$ between say $N=1000$ and some $N>>1000$ is very small and should not influence the estimation of $\mu$. 

```python
model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real owt[N,T]; // proportion of time spent in wolfpack areas
    int sel[N,T];
}
parameters {
    real<lower=0,upper=1> mu[N];
    real<lower=0,upper=100> nu[N];
    real<lower=0,upper=1> mmu;
    real<lower=0,upper=1000> nmu;
}
model {
    nmu~pareto(0.1,1.5);
    nu~pareto(0.1,1.5);
    for (n in 1:N){
        mu[n]~beta(mmu*nmu,(1-mmu)*nmu);
        for (t in 1:T){
            if (sel[n,t]==1) owt[n,t]~beta(mu[n]*nu[n],(1-mu[n])*nu[n]);
        }
    }
}
"""
man=[-0.4,-0.2,0,0.2,0.4]

G=np.zeros((D.shape[0],T,5))*np.nan
for n in range(D.shape[0]):
    for t in range(T):
        for m in range(len(man)):
            for b in range(2):
                if man[m]==D[n,b,t,10]:
                    G[n,t,m]=D[n,b,t,0]
seln=~np.all(np.isnan(G),axis=1) 
mus=[]

for m in range(len(man)):
    d=G[seln[:,m],:,m]
    if man[m]!=0:
        d.sort(1)
        d=d[:,:21]
    dat = {'N': d.shape[0],'T':d.shape[1],'owt':d,'sel':np.int32(~np.isnan(d))}
    inpar=[{'mmu':0.5,'mu':d.mean(1)}]*4
    fit = pystan.stan(model_code=model, data=dat,iter=10000, chains=4,warmup=2000,thin=5,init=inpar)
    print m,fit
    saveStanFit(fit,'E2B2WBB%d.fit'%m)
```

Next we estimate the relation between the manipulated displacement and the preference for wolfpack areas. We do this by estimating the slope of a regression line between the two variables with manipulated displacement as predictor. This is not straightforward because we do not use a model with normally distributed errors. Nevertheless, we can try to supply linear equation for the mean proportion $\mu= \alpha_0 + \alpha_1 *d$, where $d$ the is the predictor (manipulated dispalacement). The problem with this approach is that we can't ensure that the result for $\mu$ will be in the range $[0,1]$. However, we can get the sampler going by constraining $\alpha_0$ and $\alpha_1$ to particular range such that the result of the addition will be mostly in the range $[0,1]$.

We start by fitting a regression model to each subject separately and then we resume with a hierarchical model.

```python
model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real owt[N,2,T]; // proportion of time spent in wolfpack areas
    real pmd[N,2,T]; // magnitude of physical displacement
    int sel[N,2,T];
}
parameters {
    
    real<lower=0,upper=1> a0[N];
    real<lower=-0.5,upper=0.5>  a1[N];
    real<lower=0,upper=100> nu[N];
}

model {
    real mu;
    nu~pareto(0.1,1.5);
    for (n in 1:N){
    for (b in 1:2){
        for (t in 1:T){
            if (sel[n,b,t]==1){ 
                mu <- a0[n]+a1[n]*pmd[n,b,t];
                owt[n,b,t]~beta(mu*nu[n],(1-mu)*nu[n]);
            }
        }
    }}    
}
"""
dat = {'N': len(vpn),'T':T,'pmd':D[:,:,:,10],'owt':D[:,:,:,0],'sel':np.int32(~np.isnan(D[:,:,:,10]))}
fit = pystan.stan(model_code=model, data=dat,iter=10000, chains=4,warmup=2000,thin=5)
saveStanFit(fit,'E2B2Wreg.fit')
```

```python
man=[-0.4,-0.2,0,0.2,0.4]
indat=[]
for m in man:
    G=[[],[],[],[]];sels=[]
    for i in range(len(vpn)):
        if m==0: d=np.copy(D[i,0,:,:])
        else:
            d=np.copy(D[i,1,:,:])
            d[d[:,10]!=m,:]=np.nan
            ind=np.argsort(d[:,10])
            d=d[ind,:]
            d=d[:21,:]
        T=d.shape[0]
        # predictor var
        px=np.cos([np.zeros(T),d[:,7],d[:,8],d[:,9]])
        py=np.sin([np.ones(T)*np.pi/2.0,d[:,7],d[:,8],d[:,9]])
        
        # outcome var
        ox=np.cos(d[:,6])*d[:,5]
        oy=np.sin(d[:,6])*d[:,5]
        sel=d[:,4]==1
        
        if sel.mean()>0.1:
            sels.append(sel)
            G[0].append(px.T);G[1].append(py.T)
            G[2].append(ox); G[3].append(oy)
    sels=np.array(sels)
    print sels.shape, np.array(G[0]).shape
    dat = {'N': sels.shape[0],'T':sels.shape[1],'sel':np.int32(sels),'px': G[0],'py':G[1],'ox':G[2],'oy':G[3]}
    indat.append(dat)
```

```python
modelhier = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real px[N,T,4]; // predictor on x axis
    real py[N,T,4]; // predictor on y axis
    real ox[N,T]; // outcome on x axis
    real oy[N,T]; // outcome on y axis
    int<lower=0,upper=1> sel[N,T]; // indicator var to include item
}
parameters {
    real<lower=-4,upper=4> a1[N];
    real<lower=-4,upper=4> a2[N];
    real<lower=-4,upper=4> a3[N];
    real<lower=-4,upper=4> a4[N];
    real<lower=0,upper=10> sigma[N];
    real<lower=-4,upper=4>  ma1;
    real<lower=-4,upper=4>  ma2;
    real<lower=-4,upper=4>  ma3;
    real<lower=-4,upper=4>  ma4;
    real<lower=0,upper=200> sa[6];
}
model {
    real mx;
    real my;
    ma1~normal(0,2);ma2~normal(0,2);
    ma3~normal(0,2);ma4~normal(0,2);
    a1~normal(ma1,sa[1]);
    a2~normal(ma2,sa[2]);
    a3~normal(ma3,sa[3]);
    a4~normal(ma4,sa[4]);
    sigma~gamma(sa[5],sa[6]);
    for (n in 1:N){
        for (t in 1:T){
            if (sel[n,t]==1){
                mx<-a2[n]*px[n,t,2]+a3[n]*px[n,t,3]+a4[n]*px[n,t,4];
                my<-a1[n]*px[n,t,1]+a2[n]*py[n,t,2]+a3[n]*py[n,t,3]+a4[n]*py[n,t,4];
                ox[n,t] ~ normal(mx,sigma[n]);
                oy[n,t] ~ normal(my,sigma[n]);
            }
        }
    }
}
"""
for m in range(len(man)):
    fit = pystan.stan(model_code=modelhier, data=indat[m],iter=5000, chains=4,thin=2,warmup=1000)
    print fit
    saveStanFit(fit,'E2B2LH%d.fit'%m)
```
