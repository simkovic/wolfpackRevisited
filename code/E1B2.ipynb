```python
%pylab inline
%pylab --no-import-all
import numpy as np
import pylab as plt
from Evaluation import *
from  matusplotlib.plottingroutines import getColors, errorbar, pystanErrorbar
from scipy import stats

vpn=range(301,314)
D=loadDataB12(vpn)
```

In block 2 we measured how physical displacement in direction of agent's orientation (while ommiting the eyes) influences the proportion of time spent in wolfpack quadrants and perceived memory displacement.

```python
clrs=getColors(len(vpn))
plt.figure(figsize=(16,6))
perturb=np.random.rand(T)*0.02-0.01
for i in range(len(vpn)): 
    if i<5: plt.subplot(1,2,1)
    else: plt.subplot(1,2,2) 
    ppl.scatter(plt.gca(),D[i,1,:,10]+perturb,D[i,1,:,0],color=clrs[i])
    temp=[]
    for m in man: temp.append(np.median(D[i,1,D[i,1,:,10]==m,0])) 
    plt.plot(man,temp,color=clrs[i])
    plt.ylabel('Proportion of trial spent in wolfpack quadrants')
    plt.xlabel('Physical displacement [deg]')
    plt.ylim([0,1])
    plt.title('Points show trials, lines show subject averages')
```

We have split the subjects into two batches along the two different sets of manipulated variables. Though, the above plots are not helpful due to poor signal-to-noise ratio. In good faith we move directly to the slope estimation.

```python
import pystan
model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real<lower=0,upper=1> owt[N,T]; // proportion of time spent in wolfpack areas
    real<lower=-0.2,upper=0.35> pmd[N,T]; // magnitude of physical displacement
}
parameters {
    real a0[N];
    real a1[N];
    real<lower=0> asd[N];
    real ma0;
    real ma1;
    real<lower=0> sa0;
    real<lower=0> sa1;
    //real<lower=0,upper=1> theta;  
}
model {
    a0~normal(ma0,sa0);
    a1~normal(ma1,sa1);
    for (n in 1:N){
        for (t in 1:T){
            owt[n,t]~normal(a0[n]+a1[n]*pmd[n,t],asd[n]);
        }
    }
}
"""
dat = {'N': len(vpn),'T':T,'pmd':D[:,1,:,10],'owt':D[:,1,:,0]}
fit = pystan.stan(model_code=model, data=dat,iter=1000, chains=4)
print fit
pystanErrorbar(fit)
```

The estimates of individual parameters are strongly pulled towards the sample mean. The slope estimate is -0.05, which is plausible. A displacement of magnitude 0.3 (corresponding to the magnitude of Orientation Displacement of the circles with eyes) would result in avoidance of wolfpack areas with 0.485 proportion of time spent in wolfpack quadrants. However the estimates are unreliable. To obtain more reliable estimates in the next experiment we investigate only two values 0.2 and 0.4 and estimate outcome magnitude for each value alone, instead of trying to determine the functional shape.

This evaluation is mainly useful as control. Though it is possible that subjects are more sensitive to slight diversion from the motion trajectory than a veridical mechanism would suggest in this case the slope should be greater than 1. 

```python
clrs=getColors(len(vpn))
plt.figure(figsize=(16,6))
for i in range(len(vpn)):
    if i<5: plt.subplot(1,2,1)
    else: plt.subplot(1,2,2) 
    temp=[]
    phi=D[i,1,:,6]-D[i,1,:,9]
    x=np.cos(phi)*D[i,1,:,5]
    for m in man: temp.append(np.median(x[D[i,1,:,10]==m]))
    plt.plot(man,temp,color=clrs[i])
    plt.xlabel('Manipulated displacement')
    plt.ylabel('Reported displacement')
```

Again, we see mostly noise. Also, the plotted values are still confounded by lazy hand gravity and other factors. We now move to estimation analoguous to the one in block 1. 

```python
G=[[],[],[],[]];sels=[]
for i in range(len(vpn)):
    # predictor var
    px=np.cos([np.zeros(T),D[i,1,:,7],D[i,1,:,8],D[i,1,:,9]])
    py=np.sin([np.ones(T)*np.pi/2.0,D[i,1,:,7],D[i,1,:,8],D[i,1,:,9]])
    
    # outcome var
    ox=np.cos(D[i,1,:,6])*D[i,1,:,5]
    oy=np.sin(D[i,1,:,6])*D[i,1,:,5]
    sel=D[i,1,:,4]==1
    sels.append(sel)
    G[0].append(px.T);G[1].append(py.T)
    G[2].append(ox); G[3].append(oy)
sels=np.array(sels)
seln=sels.mean(1)>0.1
sels=sels[seln,:]
for i in range(seln.size):
    if not seln[i]: 
        for k in range(4): G[k].pop(i)
```

```python
import pystan
model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real<lower=-1,upper=1> px[N,T,4]; // predictor on x axis
    real<lower=-1,upper=1> py[N,T,4]; // predictor on y axis
    real<lower=-20,upper=20> ox[N,T]; // outcome on x axis
    real<lower=-20,upper=20> oy[N,T]; // outcome on y axis
    int<lower=0,upper=1> sel[N,T]; // indicator var to include item
    real<lower=-0.2,upper=0.35> pmd[N,T]; // magnitude of physical displacement
}
parameters {
    real a1[N];
    real a2[N];
    real a3[N];
    real r0[N];
    real r1[N];
    real<lower=0> sigma[N];
    real ma[5];
    real<lower=0> sa[5];
}
model {
    real mx;
    real my;
    a1~normal(ma[1],sa[1]);
    a2~normal(ma[2],sa[2]);
    a3~normal(ma[3],sa[3]);
    r0~normal(ma[4],sa[4]);
    r1~normal(ma[5],sa[5]);
    for (n in 1:N){
        for (t in 1:T){
            if (sel[n,t]==1){
                mx<-a2[n]*px[n,t,2]+a3[n]*px[n,t,3]+(r0[n]+r1[n]*pmd[n,t])*px[n,t,4];
                my<-a1[n]*py[n,t,1]+a2[n]*py[n,t,2]+a3[n]*py[n,t,3]+(r0[n]+r1[n]*pmd[n,t])*py[n,t,4];
                ox[n,t] ~ normal(mx,sigma[n]);
                oy[n,t] ~ normal(my,sigma[n]);
            }
        }
    }
}
"""
dat = {'N': seln.sum(),'T':T,'sel':np.int32(sels),'pmd':D[seln,1,:,10],'px': G[0],'py':G[1],'ox':G[2],'oy':G[3]}
fit = pystan.stan(model_code=model, data=dat,iter=1000, chains=4,thin=5)
print fit
pystanErrorbar(fit)
```

The intercept of the Orientation displacement is around zero. The slope is greater than one although there is lot of uncertainty in the estimate.
