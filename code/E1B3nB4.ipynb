Very few outliers occurs and no noticable assymetry in the histograms. The histograms across all subjects and trials show somewhat lower kurtosis, but this would be expected if the distribution are mixtures of gaussians. We will use normal distribution to model variability between trials. We fit a separate mean and std parameter for each subject.

```python

model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real dr[N,2,T]; // discrepancy
}
parameters {
    real<lower=-10,upper=10> mu[N,2];
    real<lower=0,upper=10> sigma[N,2];
    real<lower=-10,upper=10> mmu[2];
    real<lower=0,upper=10> smu[2];
}
model {  
    for (n in 1:N){
        //mu[n,1]~normal(mmu[1],smu[1]);
        //mu[n,2]~normal(mmu[2],smu[2]);
        for (t in 1:T){
            dr[n,1,t] ~ normal(mu[n,1],sigma[n,1]);
            dr[n,2,t] ~ normal(mu[n,2],sigma[n,2]);
        }
    }
}
"""
dat = {'N':seln.sum(),'T':selt.sum(),'dr':B3[seln,:2,:]}
fit = pystan.stan(model_code=model, data=dat,iter=5000,chains=4,thin=1)
print fit
plotB3fit(fit,suffix='N')
```

Let's see if medians give different results.

```python
dat = {'N':seln.sum(),'T':selt.sum(),'dr':B3r[seln,:2,:]}
fit = pystan.stan(model_code=model, data=dat,iter=5000,chains=4,thin=1)
plotB3fit(fit,suffix='NR')
print fit
```

Again we see that medians are smaller in magnitude, though they do not improve the precision of the estimates.

We now pool the estimates across subjects.

```python
model=model.replace('//mu[','mu[')
fit2 = pystan.stan(model_code=model, data=dat,iter=5000,chains=4,thin=1)
dat = {'N':seln.sum(),'T':selt.sum(),'dr':B3[seln,:2,:]}
fit1 = pystan.stan(model_code=model, data=dat,iter=5000,chains=4,thin=1)
print fit1
print fit2
plotB3fit(fit1,suffix='NH')
plotB3fit(fit2,suffix='NHR')
```

Let's see whether a model robust to outliers makes difference. We use Student's t distribution with $\nu=3$.

```python
model=model.replace('normal(mu','student_t(3,mu')
fit = pystan.stan(model_code=model, data=dat,iter=5000,chains=4,thin=1)
plotB3fit(fit,suffix='TH')
print fit
```

The results for the x axis are very similar. Note also that the model with gaussian link function has higher likelihood (though this may be the result of fixing $\nu = 3$, which may not be appropriate model for all subjects).

We now look at control trials where displacement was manipulated.

```python
plotB4(B4)
```

Points show individual trials. Lines show mean estimates for each subject. It is difficult to see any pattern because we have few observations and the expected linear effect is small. We continue by fitting regression lines to the data.

```python
model = """
data {
    int<lower=0> N; // number of subjects
    int<lower=0> T; // number of trials
    real drr[N,2,T]; // discrepancy
    real mn[N,T]; // manipulated value
}
parameters {
    real mu[N,2];
    real<lower=0,upper=10> sigma[N,2];
    real r[N,2];
    real<lower=-10,upper=10> mmu[2];
    real<lower=0,upper=10> smu[2];
    real mr[2];
    real<lower=0> sr[2];
}
model {
    
    for (n in 1:N){
        mu[n,1]~normal(mmu[1],smu[1]);
        mu[n,2]~normal(mmu[2],smu[2]);
        //r[n,1]~normal(mr[1],sr[1]);
        //r[n,2]~normal(mr[2],sr[2]);
        for (t in 1:T){
            drr[n,1,t] ~ normal(mu[n,1]+r[n,1]*mn[n,t],sigma[n,1]);
            drr[n,2,t] ~ normal(mu[n,2]+r[n,2]*mn[n,t],sigma[n,2]);
        }
    }
}
"""
dat = {'N': seln.sum(),'T':B4.shape[2],
       'drr':B4[seln,:2,:],'mn':B4[seln,2,:]}
fit = pystan.stan(model_code=model, data=dat,iter=5000, chains=4,thin=1)
print fit
plotB3fit(fit,pars=['r','mmu'],suffix='B4N')
```

We observe some subjects are outliers.

```python
modelHN=model.replace('//r[','r[')
fit = pystan.stan(model_code=modelHN, data=dat,iter=5000, chains=4,thin=1)
print fit
plotB3fit(fit,pars=['r','mr'],suffix='B4HN')
```

The estimates show very wide confidence intervals. This may be due to the outliers. Let's use t distribution to model subject variability. 

```python
modelHT=modelHN.replace('normal(mr[','student_t(3,mr[')
fit = pystan.stan(model_code=modelHN, data=dat,iter=5000, chains=4,thin=1)
print fit
plotB3fit(fit,pars=['r','mr'],suffix='B4HT')
```

The estimte of the X axis is somewhat better.
